2. Fetch around 5000 posts about cars from the Entry Level Luxury forum
https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans


```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By

options = Options()
options.log.level = "trace"  # Enable trace logs to help with debugging

ChromeDriver = webdriver.Chrome(options=options)
```

```python
# Create empty lists to hold the scraped data
commentsDictList = []
postCount = 0

for pageDecrement in range(1, 436):
    page = 436 - pageDecrement
    ChromeDriver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans/p' + str(page))
    commentBlocks = ChromeDriver.find_elements(By.XPATH, '//div[@class="Comment"]')

    # Loop through each comment block
    for commentBlock in commentBlocks:
        ChromeDriver.execute_script('''
        var blockquotes = arguments[0].querySelectorAll("blockquote");
        blockquotes.forEach(function(el) {
            el.remove();
        });
        ''', commentBlock)

        username = commentBlock.find_element(By.XPATH, './/span[contains(@class, "Author")]/a[contains(@class, "Username")]').text


        # Extract the text content
        commentText = commentBlock.find_element(By.XPATH, './/div[contains(@class, "Message userContent")]').text.replace('\n', ' ')

        # Extract the time it was posted
        commentTime = commentBlock.find_element(By.XPATH, './/span[@class="MItem DateCreated"]/a/time').get_attribute('datetime')

        commentsDict = {'userName' : username ,'Post': postCount, 'Time': commentTime, 'Text': commentText, 'page': page}

        commentsDictList.append(commentsDict)
        postCount += 1
        commentsDict = {}

    print('Page ' + str(page) + ' done')
```

```python
import pandas as pd
```

```python
commentsScraped = pd.DataFrame(commentsDictList)
```

```python
commentsScraped.to_csv('DiscussionData.csv', index=False)
```
